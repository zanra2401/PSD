# Bussiness Understanding

## Latar Belakang

Penyakit tular vektor (vector-borne diseases) masih menjadi ancaman kesehatan utama di dunia tropis, termasuk Indonesia. Masalah utamanya bukan hanya keberadaan nyamuk, tetapi spesifisitas bahaya yang dibawa oleh setiap jenis nyamuk. Kesalahan dalam mengidentifikasi jenis ancaman ini berakibat fatal pada kesalahan penanganan.

## Tujuan Bisnis

Tujuan utama proyek ini adalah membangun Early Warning System (Sistem Peringatan Dini).Deteksi Dini Wabah: 
- Mengetahui kenaikan populasi nyamuk vektor (Aedes/Culex) sebelum mereka sempat menularkan penyakit ke manusia.
- Intervensi Tepat Sasaran (Precision Public Health):
    - Jika terdeteksi Aedes (Aktif siang) $\rightarrow$ Fokus pemberantasan sarang nyamuk (3M) di genangan air bersih & Fogging pagi/sore.
    - Jika terdeteksi Quinx/Culex (Aktif malam) $\rightarrow$ Fokus pembersihan got/selokan & pembagian kelambu.
    - Jika terdeteksi Fruit Flies $\rightarrow$ Tidak perlu tindakan medis (hemat anggaran).
- Efisiensi Anggaran: Mencegah pemborosan sumber daya (racun serangga/tenaga kerja) untuk menyemprot daerah yang sebenarnya hanya dipenuhi lalat atau nyamuk jantan (tidak menggigit).

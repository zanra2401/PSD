```python
import os
import pandas as pd
import numpy as np

# list_suara_buka = os.listdir('./data_audio/buka_wav_asli')
list_suara_tutup = pd.DataFrame(os.listdir('./data_audio/tutup_wav_asli'), columns=['path'])
```

```python
def gabung_path(data):
    return './data_audio/tutup_wav_asli/' + data

list_suara_tutup = list_suara_tutup.apply(gabung_path)
list_suara_tutup.iloc(axis=0)[99]
```

```python
new_path = './data_new/tutup/new.wav'
new_path
```

## Melakukan Perbandingan Dengan DTW

```python
import librosa
import numpy as np
import matplotlib.pyplot as plt

exist_df = pd.DataFrame()

if os.path.exists("./audio_rank.csv"):
    exist_df = pd.read_csv("./audio_rank.csv").reindex(['id', 'path', 'dtw'], axis=1)

path_done = exist_df['path'].tolist() if exist_df.items() else []

audio_rank = exist_df.to_numpy().tolist() if exist_df.items() else []

new_audio, new_sr = librosa.load(new_path)
i = 0
for id, path_audio in list_suara_tutup['path'].items():
    print(i)
    if path_audio not in path_done:
        audio, sr = librosa.load(path_audio)
        dtw = librosa.sequence.dtw(new_audio, audio)[0][-1, -1]
        audio_rank.append([id, path_audio, dtw])
        new_df = pd.DataFrame(audio_rank, columns=['id', 'path', 'dtw'])
        new_df.to_csv('audio_rank.csv', index=False)
    i+= 1
```
